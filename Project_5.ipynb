```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r'C:\UNNI\Analytics\data\trial\german_credit_data.csv',index_col=0)
df.head()
```

```python
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

features = ['Age', 'Job', 'Housing', 'Saving accounts','Checking account','Credit amount','Duration']
target = 'Risk'

X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

cat_features = ['Job', 'Housing', 'Saving accounts','Checking account']

model = CatBoostClassifier(
    cat_features=cat_features,
    iterations=300,        
    depth=6,              
    learning_rate=0.1,     
    loss_function='Logloss',
    random_state=42,
    verbose=0              
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(" Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import joblib
joblib.dump(model,"model3.pkl")
```

```python
%%writefile trial_web3.py

import streamlit as st
import joblib




import streamlit as st
import pandas as pd
import joblib


model = joblib.load("model3.pkl")

st.title("Credit Risk Prediction App")
st.write("Enter applicant information to predict credit risk")




age = st.number_input("Age", min_value=18, max_value=100, value=30)

job = st.selectbox("Select Job", [1, 2, 3])
housing = st.selectbox("Housing", ["own", "rent", "free"])
saving_account=st.number_input("Saving_account_amount", min_value=0, max_value=100000, value=30000)
checking_account=st.number_input("Checking_account_amount", min_value=0, max_value=100000, value=30000)

credit_amount = st.number_input("Credit Amount", min_value=0, value=1000)
duration = st.number_input("Duration (months)", min_value=1, value=12)


df2 = pd.DataFrame({
    "Age": [age],
    "Job": [job],
    "Housing": [housing],
    "Saving accounts": [saving_account],
    "Checking account": [checking_account],
    "Credit amount": [credit_amount],
    "Duration": [duration]
})



# --- Prediction ---
if st.button("Predict Risk"):
    pred = model.predict(df2)[0]
    if pred == 1:
        st.success(" The predicted credit risk is **good**")
    else:
        st.error(" The predicted credit risk is **bad**")
```
